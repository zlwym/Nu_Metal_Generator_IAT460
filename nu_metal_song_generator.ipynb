```python
from collections import defaultdict
import random

def build_markov_chain(text, order=2):
  chain = defaultdict(list)
  words = text.split()

  for i in range(len(words) - order):
    state = tuple(words[i:i+order])
    next_word = words[i+order]
    chain[state].append(next_word)

  return chain

MOOD_WORDS = {
    "Love": {"love", "chill", "free", "grace", "bright", "joy", "dreams", "feel", "light", "feelings"},
    "Angst": {"vengeance", "hate", "hated", "anger", "death", "fear"},
    "Emotional turmoil": {"blue", "sad", "fear", "pain", "woe", "gentle"},
    "Childhood Trauma" : {"freak", "lies", "mistreated", "lonely", "empty"}
}

def generate_with_mood(chain, mood=None, num_words=5):
    words = list(random.choice(list(chain.keys())))
    order = len(words)

    for _ in range(num_words - order):
        state = tuple(words[-order:])
        if state not in chain:
            break
        candidates = chain[state]

        if mood:
            biased = []
            for w in candidates:
                if w.lower() in MOOD_WORDS.get(mood, set()):
                    biased.extend([w]*3)
                else:
                    biased.append(w)
            next_word = random.choice(biased)
        else:
            next_word = random.choice(candidates)

        words.append(next_word)
    return " ".join(words)
```

```python
import random
# Define a poetry-specific grammar
song_grammar = {
    'LYRICS': [['LINE', ',', 'LINE']],  # Four-line poem
    'LINE': [['PHRASE']],  # Two phrases per line
    'PHRASE': [
        ['PN', 'V', 'PREP', 'DET', 'ADJ', 'N'],
        ['PN', 'V', 'DET', 'N'],
        ['DET', 'ADJ', 'N', 'PREP', 'DET', 'N'],
        ['ADJ', 'N', 'ADJ', 'N'],
        ['N', 'V', 'DET', 'N']
    ],  # Phrase structure
    'ADJ': [ ],
    'N': [
          'death', 'ghost', 'pain', 'anguish', 'stone', 'box', 'blood',
          'cage', 'soul', 'freak', 'psycho', 'monster', 'poison', 'machine',
          'mind', 'enemy', 'mirror', 'heart', 'nightmare', 'memory', 'scar',
          'wound', 'breath', 'voice', 'tears', 'darkness', 'night', 'void',
          'smoke', 'abyss', 'fight', 'war', 'chain', 'prison', 'addiction'
          ],
    'V': [ ],
    'ADV': [
          'slowly', 'sadly', 'wickedly', 'viciously', 'brutally',
          'painfully', 'desperately', 'hopelessly', 'angrily',
          'coldly', 'recklessly', 'wildly', 'secretly', 'quietly'
            ],
    'PREP': [
          'in', 'at', 'on', 'of', 'to', 'by', 'with', 'above', 'between',
          'below', 'under', 'behind', 'inside', 'over', 'onto'
          ],
    'DET': [
          'the', 'a', 'this', 'that', 'his', 'her',
          'my', 'your', 'our', 'their', 'whose'
          ],
    'PN':[
        'I', 'you'
         ]
}

start_grammar = {
    'PHRASE': [
        ['I', 'was', 'PREP', 'DET', 'N', ',', 'I', 'was', 'STATE'],
    ],
    'N': [
        'bar', 'home', 'school', 'backyard'
    ],
    'DET': [
        'the', 'a', 'this', 'that', 'his', 'her',
        'my', 'your', 'our', 'their'
    ],
    'PREP': [
        'in', 'at', 'by',
    ],
    'STATE': [
        'all alone', 'with you', 'with her', 'with him'
    ]
}

MOODS = {
    'Love': {
        'ADJ': ['forbidden', 'passionate', 'romantic', 'vulnerable',
                'toxic', 'jealous', 'obsessive', 'possessive'],
        'V': ['love', 'adore', 'want', 'dream', 'break', 'crave',
              'need', 'stalk', 'spiral', 'tear', 'rip', 'destroy']
    },
    'Angst': {
        'ADJ': ['tense', 'hostile', 'cold', 'spiteful', 'ruthless',
                'bitter', 'furious', 'angry', 'violent', 'livid'],
        'V': ['hate', 'fear', 'loathe', 'rage', 'scream', 'shout',
              'smash', 'destroy', 'burn', 'explode', 'break']
    },
    'Emotional Turmoil': {
        'ADJ': ['raw', 'conflicted', 'broken', 'hollow', 'numb',
                'lifeless', 'troubled', 'isolated', 'blue'],
        'V': ['struggle', 'panic', 'mope', 'cry', 'sob', 'bleed',
              'suffer', 'ache', 'beg', 'plead', 'fall', 'battle'
              ]
    },
    'Childhood Trauma': {
        'ADJ': ['tragic', 'nostalgic', 'lonely', 'forgotten',
                'lost', 'fragile', 'abandoned', 'disturbed'],
        'V': ['mourn', 'lament', 'grieve', 'drown', 'forgive',
              'lose', 'run', 'drift', 'vanish', 'fade', 'regret'
              ]
    }
}

def make_mood_grammar(mood):
    g = song_grammar.copy()

    # Copy lists to avoid mutation bugs
    g['ADJ'] = MOODS[mood]['ADJ'][:]
    g['V'] = MOODS[mood]['V'][:]

    return g

def generate(symbol, grammar):

    if isinstance(symbol, str) and symbol in grammar:
        production = random.choice(grammar[symbol])
        if isinstance(production, list):
            return ' '.join(generate(sym, grammar) for sym in production)
        return production
    return symbol
```

```python
import random

# Grammar components

bands = [
    "Limp Bizkit",
    "Linkin Park",
    "Korn",
    "System of a Down"
]

themes = [
    "Love",
    "Angst",
    "Emotional Turmoil",
    "Childhood Trauma"
]

start_points = [
    "at high school",
    "at a bar",
    "at home",
    "all alone"
]

main_events = [
    "I got bullied",
    "I got my heart broken",
    "Mom and dad ain't treat me right",
]

outcomes = [
    "I come to terms with it",
    "I face deep emotional turmoil as a result"
]

theme = random.choice(themes)


grammar = make_mood_grammar(theme)

chorus = ''
bridge = ''

for _ in range(3):
  chorus += ' \n'
  chorus += (generate('LYRICS', grammar))

for _ in range(3):
  bridge += ' \n'
  bridge += (generate('LYRICS', grammar))


def generate_song():
    band = random.choice(bands)
    start = random.choice(start_points)

    if theme == "Love":
        main_event = "I fell in love"
    else:
        main_event = random.choice(main_events)

    outcome = random.choice(outcomes)

    return {
        "band": band,
        "start": start,

        "main_event": main_event,
        "outcome": outcome
    }


def format_lyrics(song):
    return f"""
    NU METAL SONG GENERATOR
    This generates a random song from the genre 'nu metal' based on a randomly selected theme and artist.

    {generate_with_mood(chain, mood=theme)} by {song['band']}

    Band: {song['band']}
    Theme: {theme}

    --------------------------------------------------

    {generate('PHRASE', start_grammar)}
    {generate_with_mood(chain, mood=theme, num_words = 10)}
    {generate_with_mood(chain, mood=theme, num_words=15)}
    {generate_with_mood(chain, mood=theme, num_words=10)}

    {chorus}

    {song['main_event']}
    {generate_with_mood(chain, mood=theme, num_words=10)}
    {generate_with_mood(chain, mood=theme, num_words=15)}
    {generate_with_mood(chain, mood=theme, num_words=10)}

    {chorus}

    {bridge}

    {chorus}

    {generate_with_mood(chain, mood=theme, num_words=10)}
    {generate_with_mood(chain, mood=theme, num_words=15)}
    {generate_with_mood(chain, mood=theme, num_words=10)}
    And now, {song['outcome']}.
    """


# Generate songs

song = generate_song()
print(format_lyrics(song))
```
